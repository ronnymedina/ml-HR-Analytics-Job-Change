```python
from sklearn.preprocessing import StandardScaler
# from sklearn.pipeline import make_pipeline
```

```python
cat_cols = ['enrolled_university', 'education_level', 'relevent_experience', 'experience', 'last_new_job']
num_cols = ['enrollee_id', 'training_hours', 'target']

cat_dataset_train = dataset_train[cat_cols]
num_dataset_train = dataset_train[num_cols]
cat_dataset_test = dataset_test[cat_cols]
num_dataset_test = dataset_test[['enrollee_id', 'training_hours']]
```

# Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# dataset_train_enc = pd.concat([cat_dataset_train_enc, num_dataset_train], axis=1)

sample_submission = pd.read_csv('./sample_submission.csv', encoding='utf-8')

dataset_train = pd.concat([cat_dataset_train_binenc, num_dataset_train], axis=1)
dataset_test = pd.concat([cat_dataset_test_binenc, num_dataset_test], axis=1)
dataset_test = pd.merge(dataset_test, sample_submission, on='enrollee_id', how='left')


y = dataset_train['target']
X = dataset_train.drop(['target'], axis=1)

# y_test = dataset_test['target']
# X_test = dataset_test.drop(['target'], axis=1)

# X_test_binenc = pd.concat([cat_dataset_test_binenc, num_dataset_test], axis=1)
# y_test = sample_submission = pd.read_csv('./sample_submission.csv', encoding='utf-8')

# y_train_enc = dataset_train_enc['target']
# X_train_enc = dataset_train_enc.drop(['target'], axis=1)

# y_train = dataset_train_binenc['target']
# X_train = dataset_train_binenc.drop(['target'], axis=1)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)


model = LogisticRegression()
model.fit(X, y)
y_pred = model.predict(X)
```
